# Multiple linear regression
1) Importing required libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

2) Let's import the dataset dataset.csv and explore it. Store the dataset in df. Check the structure of the data by inspecting the number of rows, columns, and attributes of the dataset. You can use functions like .shape, .head(), .info(), and .describe() to get a quick overview of the data.

```python
df= pd.read_csv('dataset.csv')
df.head()
```

3) Separate the target and the features into two variables.

```python
df.columns
```

```python
X = df[['Volume', 'Weight']]
y = df['CO2']
```

4) Plot the relationship of CO2 with other variables.

```python
sns.pairplot(df, x_vars=['Weight', 'Volume'], y_vars='CO2', height=4, aspect=1, kind='scatter')
plt.show()
```

5) Let's calculate the correlation matrix with the numerical variables in the notebook.

```python
sns.heatmap(df[['Weight', 'Volume','CO2']].corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

6) Use train_test_split to split the data into training and testing sets. Split the dataset in 80% training, 20% testing, and random_state=0.

```python
from sklearn.model_selection import train_test_split

random_state = 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

7) Create an instance of the LinearRegression and store the model in lr.

```python
from sklearn.linear_model import LinearRegression

lr =LinearRegression()
```

8) It's time to train the linear regression model using the training dataset.

```python
lr.fit(X_train,y_train)
```

9) Use the trained model to make predictions on the test data. Store the prediction in y_pred.

```python
y_pred= lr.predict(X_test)
```

10) Let's Visualize the Predictions versus the real values.

```python
plt.scatter(y_pred, y_test, color='blue', label='1:1 relation')
plt.ylabel('observed')
plt.xlabel('predicted')
plt.title('Regression analysis')
plt.legend()
plt.show()
```
