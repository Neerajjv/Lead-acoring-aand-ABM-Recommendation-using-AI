# Data Preprocessing

```python
# Converting the columns to Numeric DType as required

loan_train.Client_Income.replace("$", np.nan, inplace = True)
loan_train.Client_Income = pd.to_numeric(loan_train.Client_Income)
```

```python
loan_train.Credit_Amount.replace("$", np.nan, inplace = True)
loan_train.Credit_Amount = pd.to_numeric(loan_train.Credit_Amount)
```

```python
loan_train.Loan_Annuity.replace(["$", "#VALUE!"], np.nan, inplace = True)
loan_train.Loan_Annuity = pd.to_numeric(loan_train.Loan_Annuity)
```

```python
loan_train.Age_Days.replace("x", np.nan, inplace = True)
loan_train.Age_Days = pd.to_numeric(loan_train.Age_Days)
```

```python
loan_train.Employed_Days.replace("x", np.nan, inplace = True)
loan_train.Employed_Days = pd.to_numeric(loan_train.Employed_Days)
```

```python
loan_train.Registration_Days.replace("x", np.nan, inplace = True)
loan_train.Registration_Days = pd.to_numeric(loan_train.Registration_Days)
```

```python
loan_train.ID_Days.replace("x", np.nan, inplace = True)
loan_train.ID_Days = pd.to_numeric(loan_train.ID_Days)
```

```python
loan_train["Score_Source_3"].replace("&", np.nan, inplace=True)
loan_train.Score_Source_3 = pd.to_numeric(loan_train.Score_Source_3)
```

```python
# New Column needs to be created on the basis of Population_Region_Relative column for describing whether the Client area is
# Populated area or not

loan_train["Population_Region_Relative"].replace("#", np.nan, inplace=True)
loan_train["Population_Region_Relative"].replace("@", np.nan, inplace=True)
loan_train.Population_Region_Relative = pd.to_numeric(loan_train.Population_Region_Relative)
PRR_mean = loan_train.Population_Region_Relative.mean()
temp = loan_train.copy()
temp[temp["Population_Region_Relative"] < PRR_mean] = 0 # 0 for less populated area
temp[temp["Population_Region_Relative"] >= PRR_mean] = 1 # 1 for high populated area
loan_train["Population_Region_Relative"] = temp["Population_Region_Relative"]
```

```python
loan_train.Accompany_Client.unique()
```

```python
# Dropping Client Application ID column

loan_train.drop("ID", axis = 1, inplace=True)
```

## Null Value Analysis

```python
loan_train.info()
```

```python
from sklearn.impute import SimpleImputer

si_mean = SimpleImputer(strategy="mean", missing_values=np.nan)
si_mode = SimpleImputer(strategy="most_frequent", missing_values=np.nan)
```

```python
# Applying Mode

loan_train.iloc[:, 0 : 5] = si_mode.fit_transform(loan_train.iloc[:, 0 : 5])
loan_train.iloc[:, 8 : 16] = si_mode.fit_transform(loan_train.iloc[:, 8 : 16])
loan_train.iloc[:, 20 : 32] = si_mode.fit_transform(loan_train.iloc[:, 20 : 32])
loan_train.iloc[:, 37 : 38] = si_mode.fit_transform(loan_train.iloc[:, 37 : 38])
```

```python
# Applying Mean

loan_train.iloc[:, 6 : 8] = si_mean.fit_transform(loan_train.iloc[:, 6 : 8])
loan_train.iloc[:, 32 : 36] = si_mean.fit_transform(loan_train.iloc[:, 32 : 36])
```

```python
# Applying Mean with Round Function for req columns

loan_train.iloc[:, 5 : 6] = np.round(si_mean.fit_transform(loan_train.iloc[:, 5 : 6]))
loan_train.iloc[:, 16 : 20] = np.round(si_mean.fit_transform(loan_train.iloc[:, 16 : 20]))
loan_train.iloc[:, 36 : 37] = np.round(si_mean.fit_transform(loan_train.iloc[:, 36 : 37]))
```

```python
loan_train.Accompany_Client.unique()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

cols = ["Accompany_Client", "Client_Marital_Status", "Client_Income_Type", "Client_Education", "Client_Gender", "Loan_Contract_Type", "Client_Housing_Type", "Client_Occupation", "Client_Permanent_Match_Tag", "Client_Contact_Work_Tag", "Type_Organization"]
loan_train[cols] = loan_train[cols].apply(LabelEncoder().fit_transform)
```

```python
loan_train.Accompany_Client.unique()
```

# Model Training

```python
X = loan_train.drop("Default", axis = 1)
y = loan_train.Default
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python

y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred) * 100
```
