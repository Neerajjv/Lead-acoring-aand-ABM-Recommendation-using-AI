# Training, Saving, and Making Predictions with a Machine Learning Model
## 1) Importing required libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from joblib import dump, load
```

2) Let's import the dataset icecreamsales.csv and explore it. Store the dataset in df. Check the structure of the data by inspecting the number of rows, columns, and attributes of the dataset. You can use functions like .shape, .head(), .info(), and .describe() to get a quick overview of the data.

```python
df = pd.read_csv('Ice Cream.csv')
```

```python
df.head()
```

3) Split the target and features

```python
X = df[['Temperature', 'Advertisement']]  # Features
y = df['Sales']  # Target
```

4) Use train_test_split to split the data into training and testing sets. Split the dataset in 80% training, 20% testing, and random_state=0.

```python
from sklearn.model_selection import train_test_split

random_state = 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

5) Create an instance of the LinearRegression and store the model in lr.

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

6) It's time to train the linear regression model using the training dataset.

```python
lr.fit(X_train, y_train)
```

7) Use the trained model to make predictions on the test data. Store the prediction in y_pred.

```python
y_pred = lr.predict(X_test)
```

8) Let's Visualize the Predictions versus the real values.

```python
plt.scatter(y_pred, y_test, color='blue', label='1:1 relation')
plt.ylabel('oberved')
plt.xlabel('predicted')
plt.title('Regression analysis')
plt.legend()
plt.show()
```

9)Save the trained model to a file.

```python
dump(lr, 'linear_regression_model.joblib')
```

10) Now your Linear Regression model is saved as a file named "linear_regression_model.joblib". You can load it back later using the load function:

```python
loaded_model = load('linear_regression_model.joblib')
```
